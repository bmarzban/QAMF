## 2.3 A Coupled Electrochemical Process: ATP Synthesis 

The electrostatic energy potential across the mitochondrial inner membrane is used to drive the synthesis of ATP in the final step of oxidative ATP synthesis. The mammalian mitochondrial $\rm F_{1} F_{0} \,$ ATP synthase synthesizes ATP in the mitochondrial matrix from ADP and inorganic phosphate, coupled to the translocation of protons (H+ ions) from outside to inside of the matrix. The overall reaction to describe this process takes the following form:


\begin{equation}
\rm MgADP^{1-} + HPO4^{2-} + \frac{1}{2} O_{2} + H^{+}_{inside} + NADH \leftrightharpoons MgATP^{2-} + H_{2}O
+ NAD^{+} \label{eq}\tag{2.1}
\end{equation}


where the term $\rm H^{+}_{inside}$ indicates that a hydrogen ion from inside the matrix is covalently incorporated into the synthesized ATP. The species $\rm MgADP^{1-}$ and $\rm MgATP^{2-}$ are the magnesium-bound species of ADP and ATP. This chemical reaction is coupled to the net transport of nA = $\rm \frac{8}{3}$ protons across the inner membrane:


\begin{equation}
\rm 8(H^{+})_{outside} \leftrightharpoons 3(H^{+})_{inside}
\end{equation}

\begin{equation}
\rm nA = \frac{(H^{+})_{outside}}{(H^{+})_{inside}} = \frac{8}{3}
\end{equation}

where nA represents the P/O ratio, that is, the average number of ATP synthesized per $\rm \frac{1}{2}O_{2}$ consumed. It is important to note that this approximate P/O ratio is measured when NADH is the electron donor as shown in equation 2.1. However, when electrons are oxidized from succinate as the substrate, the average P/O ratio is reduced to $\frac{3}{2}$. This decrease in P/O ratio is measured because electrons derived from succinate oxidation bypass a proton translocation site, complex I, and instead enter through complex II.

### 2.3.1 Given a free magnesium concentration [Mg$\rm^{2+}$] and hydrogen ion activity [H$\rm^{+}$] = 10$\rm^{-pH}$, how can you compute the concentrations of MgADP$\rm^{1-}$, MgATP$\rm ^{2-}$, and HPO4$\rm^{2-}$ in terms of the total concentrations of the reactants [$\Sigma$ADP],  [$\Sigma$ATP], and [$\Sigma$Pi]? (You will need to account for binding of biochemical species to [Mg$\rm ^{2+}$] and [H$\rm ^{+}$].)

Let's start with looking at the species of ATP:

\begin{eqnarray} 
[\Sigma \rm ATP] &=& [\rm ATP^{4-}] + [\rm H \cdot \rm ATP^{3-}] + [\rm K \cdot \rm ATP^{3-}] + [\rm Mg \cdot \rm ATP^{2-}]\\
&=& [\rm ATP^{4-}] + \frac{[\rm H^{+}]}{\rm HATP^{3-}}[\rm ATP^{4-}] + \frac{[\rm K^{+}]}{\rm KATP^{3-}}[\rm ATP^{4-}] + \frac{[\rm Mg^{2+}]}{\rm MgATP^{2-}}[\rm ATP^{4-}]\\
&=& [\rm ATP^{4-}] \left(1 + \frac{[\rm H^{+}]}{\rm K_{HATP}}+ \frac{[\rm K^{+}]}{\rm K_{\rm KATP}} + \frac{[\rm Mg^{2+}]}{\rm K_{\rm MgATP}}\right)
\end{eqnarray}

(This expression ignores the two cation bound forms)

then multiplying the right hand side of the expression by $\frac{[\rm Mg^{2+}]}{\rm K_{\rm MgATP}}$, we get the following:

\begin{eqnarray}
&=& \frac{[\rm Mg^{2+}]}{\rm K_{\rm MgATP}}[\rm ATP^{4-}] \cdot 
\left(\frac{\rm K_{\rm MgATP}}{\rm[Mg^{2+}]}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [H^{+}]}{\rm K_{\rm KHATP}}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [K^{+}]}{\rm K_{\rm KATP}}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [Mg^{2+}]}{\rm K_{\rm MgATP}}\right);  
\end{eqnarray}

and given that

 $$
\rm [Mg \cdot ATP^{2-}] = \frac{\rm [Mg^{2+}]}{\rm K_{\rm MgATP}} \cdot \rm [ATP^{4-}],
 $$
 
 we substitute it into the expression and get the following:

\begin{eqnarray}
 &=& \rm [Mg \cdot ATP^{2-}] \cdot 
\left(\frac{\rm K_{\rm MgATP}}{\rm[Mg^{2+}]}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [H^{+}]}{\rm K_{\rm KHATP}}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [K^{+}]}{\rm K_{\rm KATP}}
+ \frac{\rm K_{\rm MgATP}}{\rm [Mg^{2+}]} \cdot \frac{\rm [Mg^{2+}]}{\rm K_{\rm MgATP}}\right); 
\end{eqnarray}

then pulling out the common factor $\rm \frac{K_{MgATP}}{[Mg^{2+}]}$ we arrive at the following:

\begin{eqnarray}
 &=& \rm [Mg \cdot ATP^{2-}] \cdot \frac{K_{MgATP}}{Mg^{2+}} 
\left(1
+ \frac{\rm [H^{+}]}{\rm K_{\rm KHATP}}
+ \frac{\rm [K^{+}]}{\rm K_{\rm KATP}}
+ \frac{\rm [Mg^{2+}]}{\rm K_{\rm MgATP}}\right); 
\end{eqnarray}

Then we substitute in the expresssion for the binding polynomial to account for interactions with biochemical species including $\rm H^{+},$ $\rm K^{+},$ and $\rm Mg^{2+}$ ions

$$\rm P_{ATP} = \left(1 + \frac{[H^{+}]}{K_{HATP}} + \frac{[K^{+}]}{K_{KATP}} + \frac{[Mg^{2+}]}{K_{MgATP}}\right),$$

and arrive at the following:

\begin{eqnarray}
\rm[\Sigma ATP] &=& \rm [Mg \cdot ATP^{2-}] \frac{\rm K_{\rm MgATP}}{[\rm Mg^{2+}]} \cdot 
P_{ATP}
\end{eqnarray}

which can now be solved as

$$ 
\rm [Mg \cdot ATP^{2-}] = [\Sigma ATP] \cdot \frac{[Mg^{2+}]/K_{MgATP}}{P_{ATP}}
$$


Following the same process for the other energetic substrates produces the following solutions:

\begin{eqnarray}
[\rm Mg \cdot ADP^{-}] &=& [\Sigma \rm ADP] \cdot \frac{[\rm Mg^{2+}]/K_{MgADP}}{\rm P_{\rm MgADP}}\\
[\rm HPO_{4}^{2-}] &=& [\Sigma \rm Pi] \cdot \frac{[\rm H^{+}]/K_{HPO4}}{P_{Pi}}
\end{eqnarray}


### 2.3.2 Derive an expression for the Gibbs free energy change associated with reaction (2.1) in terms of the reference $\rm \Delta G^{o}$, the concentrations of biochemical reactants, and the cation concentrations [Mg$^{2+}$] and [H$^{+}$]. What is the free energy of ATP synthesis in the mitochondrial matrix? Assume that pH = 7.2 and [$\Sigma$ADP] = 9.5 mM, [$\Sigma$ATP] = 5 mM, and [$\Sigma$Pi] = 1 mM.

Consider the ATP synthesis reaction

\begin{equation}
\rm ADP^{3-} + HPO4^{2-} + \frac{1}{2} O_{2} + H^{+}_{inside} \leftrightharpoons ATP^{4-} + H_{2}O,
\end{equation}

With equlibrium mass action ratio

\begin{equation}
\rm K_{eq} = e^{-\Delta G_{r}^{o}/RT}
= \left(\frac{[ATP^{4-}]}{[H^{+}][HPO_{4}^{2-}][ADP^{3-}]}\right)_{eq}.
\end{equation}

Substituting the expressions for the binding polynomials results in

\begin{equation}
\rm K_{eq} = e^{-\Delta G_{r}^{o}/RT}
= \left(\frac{[\Sigma ATP]}{[\Sigma ADP][\Sigma Pi]}\right)_{eq} \cdot \frac{P_{ADP}P_{Pi}}{[H^{+}]P_{ATP}},
\end{equation}

which yields an expression for the apparent equilibrium constant in terms of biochemical reactants

$$
\rm K^{'}_{eq} = e^{-\Delta_{r} G^{'o}/RT} 
= \left(\frac{[\Sigma ATP]}{[\Sigma ADP] [\Sigma Pi]}\right)_{eq}
= K_{eq} \frac{[H^{+}]P_{ATP}}{P_{ADP} P_{Pi}}.
$$

Using these equations, there are a couple of equivalent ways to express the Gibbs free energy change for this reaction. One is based on the reference chemical reaction

$$
\rm \Delta_{r} G^{'} = \Delta_{r} G^{o} + RTln \left(\frac{[ATP^{4-}]}{[H^{+}][HPO_{4}^{2-}][ADP^{3-}]}\right).
$$

The other is

$$
\rm \Delta_{r} G^{'} = \Delta_{r} G^{'o} + RTln \left(\frac{[\Sigma ATP]}{[\Sigma ADP] [\Sigma Pi]}\right).
$$

To verify that these two equations give the same value, take the following values for dissociation constants

|Reactant (L)| $\rm K_{H-L}$            | $\rm K_{K-L}$             | $\rm K_{Mg-L}$           |
|------------|--------------------------|---------------------------|--------------------------|
|ATP         |2.757 $\times$ 10$^{-7}$ M|9.809 $\times$ 10$^{-2}$ M |8.430 $\times$ 10$^{-5}$ M|
|ADP         |4.106 $\times$ 10$^{-7}$ M|1.319 $\times$ 10$^{-1}$ M |7.149 $\times$ 10$^{-4}$ M|
|Pi          |2.308 $\times$ 10$^{-7}$ M|3.803 $\times$ 10$^{-1}$ M |2.815 $\times$ 10$^{-2}$ M|

and $\rm \Delta_{r} G^{o}$ = -4.67 $\rm \frac{kJ}{mol}$, which yields $\rm K_{eq}$ = 6.11698 at 37$\rm^{o}$C.

The Python code below verifies that both approaches yield the same value 

```python
# %%

import numpy as np

# Dissociation Constants
K_HATP = 2.757e-7
K_KATP = 9.809e-2
K_MgATP = 8.430e-5
K_HADP = 4.106e-7
K_KADP = 1.319e-1
K_MgADP = 7.149e-4
K_HPI  = 2.308e-7
K_KPI  = 3.803e-1
K_MgPI  = 2.815e-2

# Other constants
DGr0 = -4670 #J/mol
RT = 310.15*8.3145 #J/mol
Keq = np.exp(-DGr0/RT)

# Concentrations
ATP = 0.5e-3
ADP = 9.5e-3
Pi  = 1e-3
Mg  = 1e-3
H   = 10**(-7.2)
K   = 100e-3

# binding polynomials:
P_ATP = 1 + H/K_HATP + K/K_KATP + Mg/K_MgATP
P_ADP = 1 + H/K_HADP + K/K_KADP + Mg/K_MgADP
P_Pi =  1 + H/K_HPI + K/K_KPI + Mg/K_MgPI

# reference species concentration
ATP_s = ATP/P_ATP
ADP_s = ADP/P_ADP
Pi_s = Pi/P_Pi

# compute DeltaG_r in terms of reference species
DG1 = +DGr0 + RT*np.log(ATP_s/(H*Pi_s*ADP_s))
print("First equation:",DG1," J/mole")

K_apparent = Keq*(H*P_ATP/(P_ADP*P_Pi))
DG2 = -RT*np.log(K_apparent) + RT*np.log(ATP/(ADP*Pi))
print("Second equation:",DG2," J/mole")
```

The positive $\rm \Delta_{r} G$ indicates that the reaction is thermodynamically unfavorable in the direction toward ATP synthesis. This result implies that ATP synthesis can only occur when coupled to some other thermodynamically favorable reaction.

### 2.3.4 Given the concentrations assumed in 2.3.2, what is the minimum value of $\rm \Delta\Psi$ at which ATP can be synthesized in the mitochondrial matrix?

At equilibrium, $\rm \Delta G = 0$, where the electrical energy consumed is equal to the chemical energy required for ATP synthesis:

\begin{equation} 
    \rm \Delta G_{\rm{ATP}} = \rm n_{A} \Delta G_{\rm{H}^{+}} = \rm n_{A} (F \Delta \Psi)
\end{equation}

Rearranging the equation to solve for the electrical energy

\begin{equation}    
\rm \Delta \Psi  = \rm \frac{\Delta G_\mathrm{ATP} }{ n_A F } = \rm 177.7 \ \mathrm{mV}.
\end{equation} 

Therefore, $\Delta \Psi$ = 177.7 is the minimum value of membrane potential at which ATP can be synthesized.


### 2.3.5 Assume that reaction (2.1) proceeds by simple mass-action kinetics, with a constant reverse rate $\rm k_{r}$. How does the forward rate constant necessarily depend on $\rm \Delta \rm \Psi$ for the reaction kinetics to be properly thermodynamically balanced?

The reaction rate kinetics can be written in terms of ATP synthesis as follows:

\begin{equation}
\rm J = J_f - J_r = k_{f}[ADP^{3-}][HPO_{4}^{2-}] - k_{r}[ATP^{4-}],
\end{equation}

or in terms of the sums of species, $\rm J_{F1F0ATPase}$

\begin{equation}
\rm J_{F1F0ATPase} = J_f - J_r = k_{f}[\Sigma ADP][\Sigma Pi] - k_{r}[\Sigma ATP].
\end{equation}

At thermodynamic equilibrium, the equation can be written as,

$$
\rm k_{f}[\Sigma ADP][\Sigma Pi] - k_{r}[\Sigma ATP]= 0.
$$

Solving for $\rm K^{'}_{F1F0ATPase}$ results in,

$$
\rm K^{'}_{F1F0ATPase} = \left(\frac{[\Sigma ATP]}{[\Sigma ADP][\Sigma Pi]}\right)_{eq} = \frac{k_{f}}{k_{r}.}
$$

Writing $\rm k_{f}$ in terms of $\rm k_{r}$ and substituting into the equation above provides,

$$
\rm k_{r}k^{'}_{F1F0ATPase}[\Sigma ADP][\Sigma Pi] - k_{r}[\Sigma ATP].
$$

Then pulling out $\rm k_{r}$ as a common factor, 

$$
\rm k_{r}(k^{'}_{F1F0ATPase}[\Sigma ADP][\Sigma Pi] - [\Sigma ATP]).
$$

Therefore, a simple mass-action expression for the reaction flux that satisfies this thermodynamic relation is

\begin{equation}
\rm J_{F1F0ATPase} = \rm k_{r}(k^{'}_{F1F0ATPase}[\Sigma ADP][\Sigma Pi] - [\Sigma ATP]),
\end{equation}

where

\begin{eqnarray} 
  \frac{\mathrm{d}}{\mathrm{dt}} [\Sigma \mathrm{ATP}] &=& J_{\mathrm{F}_1\mathrm{F}_0\mathrm{ATPase}} \\ 
  \frac{\mathrm{d}}{\mathrm{dt}} [\Sigma \mathrm{ADP}] &=& -J_{\mathrm{F}_1\mathrm{F}_0\mathrm{ATPase}} \\
  \frac{\mathrm{d}}{\mathrm{dt}} [\Sigma \mathrm{Pi}] &=& -J_{\mathrm{F}_1\mathrm{F}_0\mathrm{ATPase}} \\ 
\end{eqnarray}

#### The derivation for $\rm k^{'}_{F1F0ATPase}$ is shown below:

To recap, the free energy for the ATP synthesis reaction

\begin{equation}
\rm ADP^{3-} + \rm HPO_{4}^{2-} + \rm \frac{1}{2} O_{2} + \rm H^{+}_{inside} \leftrightharpoons \rm ATP^{4-} + \rm H_{2}O,
\end{equation}

is

\begin{equation}
\rm \Delta_{r}G_{ATP} = \rm \Delta_{r}G_{ATP}^{'o} + RTln \left(\frac{[ATP^{4-}]}{[H^{+}][HPO_{4}^{2-}][ADP^{3-}]}\right).
\end{equation}

where $\rm \Delta_{r}G_{ATP}^{o} \approx$ -4.6 kJ/mol is the standard Gibbs free energy change for the reference chemical reaction. In mitochondria, ATP synthesis is coupled to proton transport from outside of to inside of the matrix:

\begin{equation}
\rm 8(H^{+})_{outside} \leftrightharpoons 3(H^{+})_{inside}
\end{equation}

\begin{equation}
\rm nA = \frac{(H^{+})_{outside}}{(H^{+})_{inside}} = \frac{8}{3}
\end{equation}

Which is associated with a free energy change,

\begin{equation}
\rm \Delta_{r} G_{H^+} = \rm n_{A}F \Delta \Psi + \rm RTlog \left( \frac{[H^{+}]_{inside}}{[H^{+}]_{outside}}\right)^{n_{A}}.
\end{equation}

This free energy is sometimes called the proton motive force. The last term on the right-hand side represents  the proton-transport free energy produced from the pH gradient across the inner mitochondrial membrane.

The free energy of coupled ATP synthesis and proton transport is

\begin{equation}
\rm \Delta_{r} G = \rm \Delta_{r}G_{ATP}^{'o} + \rm RT log \left(\frac{[ATP^{4-}]}{[H^+][HPO_{4}^{2-}][ADP^{3-}]}\right) - \rm n_{A}F\Delta\Psi - \rm RTlog \frac{([H^{+}]_{outside})^{n_{A}}}{([H^{+}]_{inside})^{n_{A}-1}}. 
\end{equation}

Chemical equilibrium occurs when $\rm \Delta_r G = 0$:

$$
\rm RTlog \left( \frac{[HPO_{4}^{2-}][ADP^{3-}]}{[ATP^{4-}]}\right)_{eq} = \Delta_{r}G_{ATP}^{o} - n_{A}F\Delta\Psi - RTlog \frac{([H^{+}]_{outside})^{n_{A}}}{([H^{+}]_{inside})^{n_{A}-1}}. 
$$

Dividing by RT on both sides and re-writing the equilibrium expression in terms of ATP synthesis yields

$$
\rm -log \left( \frac{[ATP^{4-}]}{[HPO_{4}^{2-}][ADP^{3-}]}\right)_{eq}= \frac{\Delta_{r}G_{ATP}^{o} - n_{A}F\Delta\Psi}{RT} - log\frac{([H^{+}]_{outside})^{n_{A}}}{([H^{+}]_{inside})^{n_{A}-1}}. 
$$

Cancelling the log terms by raising both sides to the base root of 10 produces

$$
\rm k_{eq} = \left( \frac{[ATP^{4-}]}{[HPO_{4}^{2-}][ADP^{3-}]}\right)_{eq} = 10^{-\frac{(\Delta_{r}G_{ATP}^{o} - n_{A}F\Delta\Psi)}{RT}} \cdot \frac{([H^{+}]_{outside})^{n_{A}}}{([H^{+}]_{inside})^{n_{A}-1}}.
$$

Or in terms of sums of species

$$
\rm K^{'}_{F1F0ATPase} = \left(\frac{[\Sigma ATP]}{[\Sigma Pi][\Sigma ADP]}\right)_{eq} =10^{-\frac{(\Delta_{r}G_{ATP}^{o} - n_{A}F\Delta\Psi)}{RT}} \cdot \frac{P_{ATP}}{P_{ADP}P_{Pi}} \cdot \frac{([H^{+}]_{outside})^{n_{A}}}{([H^{+}]_{inside})^{n_{A}-1}}.  
$$

### 2.3.6 Write a simple program that simulates the kinetics of [$\Sigma$ADP], [$\Sigma$ATP] , and [$\Sigma$Pi] in the matrix given a fixed membrane potential, pH, and magnesium concentration, and given arbitrary initial conditions. How do the predicted steady-state concentrations depend on membrane potential, pH, and magnesium concentration?

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Constants

R = 8.3145        # J / K / mol
T = 310.15        # K
RT = R*T
F = 96485.33      # C / mol 
pH_e = 7.2
pH_x = 7.4
H_e = 10**(-pH_e) # M
H_x = 10**(-pH_x) # M
Mg = 1e-3         # M
K = 100e-3        # M
psi = 180 / 1000     # V

# Dissociation Constants

K_HATP = 2.757e-7
K_KATP = 9.809e-2
K_MATP = 8.430e-5
K_HADP = 4.106e-7
K_KADP = 1.319e-1
K_MADP = 7.149e-4
K_HPI  = 2.308e-7
K_KPI  = 3.803e-1
K_MPI  = 2.815e-2

# Binding Polynomials

P_ATPx = 1 + H_x/K_HATP + Mg/K_MATP #P_ATPx = 1 + H_x/K_HATP + K/K_KATP + Mg/K_MATP
P_ADPx = 1 + H_x/K_HADP + Mg/K_MADP #P_ADPx = 1 + H_x/K_HADP + K/K_KADP + Mg/K_MADP
P_PIx = 1 + H_x/K_HPI + Mg/K_MPI    #P_PIx = 1 + H_x/K_HPI + K/K_KPI + Mg/K_MPI

# Species
ATPx = 0.5e-3      # M
ADPx = 9.5e-3      # M
Pix  = 1e-3        # M
```

### F1F0 ATP synthase

```python
def J_F1F0(psi, pH_inner, pH_outer, Mg, concentrations):
    ATPx, ADPx, Pix  = concentrations # define initial conditions 
    
    k_r = 1
    n_H = 8/3
    DGr0 = -4670
    
    keq_prime_F1F0 = np.exp(-(DGr0 - F * psi * n_H)/(RT))*P_ATPx/(P_ADPx*P_PIx)*H_e**n_H / H_x**(n_H-1) 
    J = k_r*(keq_prime_F1F0 * Pix * ADPx - ATPx)

    return J
```

```python
init_conditions = np.array([ATPx, ADPx, Pix])
def dSTATES_dt(STATES, t, psi, pH_inner, pH_outer, Mg):
    rate = J_F1F0(psi, pH_inner, pH_outer, Mg, STATES)
    dATPx = rate
    dADPx = - rate
    dPix = - rate

    dxdt = (dATPx, dADPx, dPix)
    return dxdt

t = np.linspace(0,5)

dSTATES_dt(init_conditions, t, psi, pH_inner, pH_outer, Mg)

ode_results = odeint(dSTATES_dt,init_conditions, t, (psi, pH_inner, pH_outer, Mg)).T
plt.plot(t,ode_results[0] * 1000, label = 'ATPx')
plt.plot(t,ode_results[1] * 1000, label = 'ADPx')
plt.plot(t,ode_results[2] * 1000, label = 'Pix')

plt.xlabel('Time (s)', fontsize=16)
plt.ylabel('Concentation (mM)', fontsize=16)

plt.legend()
ode_results[:,-1][0]/ode_results[:,-1][1] 
```
